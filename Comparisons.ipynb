```python
import matplotlib.pyplot as plt

# Algorithm names
algorithms = ['Decision Tree', 'Random Forest', 'Linear Regression', 'Support Vector Machine']

# Corresponding accuracies
accuracies = [98.51, 98.13, 80.53, 42.14]

# Improved color combination
colors = ['#3498db', '#2ecc71', '#e74c3c', '#f39c12']

# Plotting the bar graph with better color combination
plt.figure(figsize=(10, 6))
plt.bar(algorithms, accuracies, color=colors)

plt.title('Algorithm Accuracies Comparison')
plt.xlabel('Algorithms')
plt.ylabel('Accuracy (%)')
plt.ylim(0, 100)  # Setting y-axis limit for better visualization
plt.show()
```

```python
import matplotlib.pyplot as plt

# Algorithm names
algorithms = ['Decision Tree', 'Random Forest', 'Linear Regression', 'Support Vector Machine']

# Actual accuracy scores
accuracy_70_30 = [73.5, 82.5, 70.3, 68.9]  # Replace with actual scores
accuracy_80_20 = [78.2, 85.6, 80.5, 76.8]  # Replace with actual scores
accuracy_75_25 = [76.4, 81.2, 78.6, 73.2]  # Replace with actual scores

# Width of the bars
bar_width = 0.2

# Create subplots for each split ratio
fig, ax = plt.subplots(figsize=(12, 6))

# Plotting the bar graphs for each split
bar_positions_70_30 = range(len(algorithms))
bar_positions_80_20 = [pos + bar_width for pos in bar_positions_70_30]
bar_positions_75_25 = [pos + bar_width * 2 for pos in bar_positions_70_30]

ax.bar(bar_positions_70_30, accuracy_70_30, width=bar_width, label='70-30 Split')
ax.bar(bar_positions_80_20, accuracy_80_20, width=bar_width, label='80-20 Split')
ax.bar(bar_positions_75_25, accuracy_75_25, width=bar_width, label='75-25 Split')

# Labeling and styling
ax.set_title('Algorithm Accuracies Comparison for Different Split Ratios')
ax.set_xlabel('Algorithms')
ax.set_ylabel('Accuracy (%)')
ax.set_xticks([pos + bar_width for pos in bar_positions_70_30])
ax.set_xticklabels(algorithms)
ax.legend()

# Set y-axis limit from 0 to 100
ax.set_ylim(0, 100)

# Show the plot
plt.show()
```
