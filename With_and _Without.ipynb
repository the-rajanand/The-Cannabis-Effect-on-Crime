```python
import pandas as pd

# Read the data from the CSV file
df = pd.read_csv("D:\Work\DMPA Mini Project\Cannabis_Dataset.csv") 

# Convert columns to numeric 
df['No dispensaries '] = pd.to_numeric(df['No dispensaries '], errors='coerce')
df['Has dispensaries'] = pd.to_numeric(df['Has dispensaries'], errors='coerce')

# Drop rows with NaN values
df = df.dropna(subset=['No dispensaries ', 'Has dispensaries'])

# Calculate sums
blocks_with_dispensaries = df['Has dispensaries'].sum()
blocks_without_dispensaries = df['No dispensaries '].sum()

# Calculate percentages
percentage_blocks_with_dispensaries = (blocks_with_dispensaries / (blocks_with_dispensaries + blocks_without_dispensaries)) * 100
percentage_blocks_without_dispensaries = (blocks_without_dispensaries / (blocks_with_dispensaries + blocks_without_dispensaries)) * 100

# Print the results
print(f"Percentage of blocks with dispensaries: {percentage_blocks_with_dispensaries}%")
print(f"Percentage of blocks without dispensaries: {percentage_blocks_without_dispensaries}%")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Convert columns to numeric
df['No dispensaries '] = pd.to_numeric(df['No dispensaries '], errors='coerce')
df['Has dispensaries'] = pd.to_numeric(df['Has dispensaries'], errors='coerce')

# Drop rows with NaN values
df = df.dropna(subset=['No dispensaries ', 'Has dispensaries'])

# Iterate over each area and create a separate bar graph
for area in df['Blocks']:
    area_df = df[df['Blocks'] == area]

    # Set the figure size
    plt.figure(figsize=(5, 3))

    # Set bar width and position
    bar_width = 0.3
    r1 = range(len(area_df['Blocks']))
    r2 = [x + bar_width for x in r1]

    # Create bars for 'Without Marijuana'
    plt.bar(r1, area_df['No dispensaries '], color='skyblue', width=bar_width, edgecolor='grey', label='Without Marijuana')

    # Create bars for 'With Marijuana'
    plt.bar(r2, area_df['Has dispensaries'], color='lightgreen', width=bar_width, edgecolor='grey', label='With Marijuana')

    # Add labels and title
    plt.xlabel('Blocks')
    plt.ylabel('Count')
    plt.title(f'Dispensaries in Area: {area}')

    # Add legend
    plt.legend()

    # Set x-axis labels
    plt.xticks([r + bar_width/2 for r in r1], area_df['Blocks'], rotation=0)  # Set rotation to 0 for horizontal labels

    # Show the plot for each area
    plt.tight_layout()
    plt.show()
```
