```python
import pandas as pd

data = {
    'Block': ['Marijuana dispensaries', 'Average household income', 'Percent homeowners', 
              'Percent vacant units', 'Proportion industrial use', 'Proportion office use', 
              'Proportion residential use', 'Proportion retail use', 'Bars', 'Liquor stores'],
    'Homicide': [1.2380, -0.0094, -0.0056, 0.0219, -0.2373, 0.4426, -0.0423, 0.0496, 0.5500, 0.0765],
    'Robbery': [0.4011, -0.0114, -0.0102, 0.0075, 0.1100, 0.3633, -0.0641, 0.1248, 0.3539, 0.3501],
    'Assault': [0.3657, -0.0138, -0.0071, 0.0130, -0.0152, 0.3036, -0.0487, -0.0758, 0.4954, 0.2828],
    'Burglary': [0.0842, -0.0014, -0.0018, 0.0011, 0.0028, 0.3226, 0.0501, 0.1442, 0.1535, 0.0959]
}

crime_data = pd.DataFrame(data)

# Calculate the total number of crimes in each block
crime_data['TotalCrimes'] = crime_data[['Homicide', 'Robbery', 'Assault', 'Burglary']].sum(axis=1)

# Calculate the percentage of each type of crime in each block
for crime_type in ['Homicide', 'Robbery', 'Assault', 'Burglary']:
    crime_data[f'{crime_type}Percentage'] = (crime_data[crime_type] / crime_data['TotalCrimes']) * 100

# Print the resulting DataFrame with percentages
print(crime_data)
```

```python
import matplotlib.pyplot as plt

# Define the data
blocks = crime_data['Block']
homicide = crime_data['Homicide']
robbery = crime_data['Robbery']
assault = crime_data['Assault']
burglary = crime_data['Burglary']

# Set up the figure and axis
fig, axs = plt.subplots(4, 1, figsize=(10, 20))

# Plot Homicide
axs[0].bar(blocks, homicide, color='red')
axs[0].set_title('Homicide')
axs[0].set_ylabel('Coefficient Value')
axs[0].tick_params(axis='x', labelrotation=45)  
# Plot Robbery
axs[1].bar(blocks, robbery, color='blue')
axs[1].set_title('Robbery')
axs[1].set_ylabel('Coefficient Value')
axs[1].tick_params(axis='x', labelrotation=45)  

# Plot Assault
axs[2].bar(blocks, assault, color='green')
axs[2].set_title('Assault')
axs[2].set_ylabel('Coefficient Value')
axs[2].tick_params(axis='x', labelrotation=45)  

# Plot Burglary
axs[3].bar(blocks, burglary, color='purple')
axs[3].set_title('Burglary')
axs[3].set_ylabel('Coefficient Value')
axs[3].tick_params(axis='x', labelrotation=45)  

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```
